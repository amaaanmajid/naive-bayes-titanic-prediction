```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('titanic.csv')
df
```

```python
df.drop(['Embarked','PassengerId','Cabin'],axis='columns',inplace=True)
df
```

```python
dummies=pd.get_dummies(df.Sex)
dummies
```

```python
a=pd.concat([df,dummies],axis='columns')
a
```

```python
a.columns[a.isna().any()]
```

```python
a.Age=a.Age.fillna(a.Age.mean())
a
```

```python
yeinput=a
yeinput
```

```python
output=df.Survived
```

```python
output
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(yeinput,output,test_size=0.2)
```

```python
len(X_train)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
model.predict_proba(X_test)
```
